```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
a = np.arange(-5,5,0.01)
b = a
a,b = np.meshgrid(a,b)
print(a)
```

```python
print(b)
print(type(b))
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a,b,a**2+b**2,cmap = 'rainbow')
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.contour(a,b,a**2+b**2,cmap = 'coolwarm')
plt.show()
```
