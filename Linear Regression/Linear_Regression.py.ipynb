## Algorithm - Linear Regression

```python
def hypothesis(theta,x):
    return theta[0]+theta[1]*x
```

```python
def error(X,Y,theta):
    m = x.shape[0]
    err = 0;
    for i in range(m):
        err+=(hypothesis(theta,X[i])-Y[i])**2
    return err/(2)
```

```python
def gradient(theta,X,Y):
    grad = np.array([0.0,0.0])
    m = x.shape[0]
    err1 = 0
    err2 = 0;
    for i in range(m):
        err1 += (hypothesis(theta,X[i])-Y[i])
        err2 += (hypothesis(theta,X[i])-Y[i])*X[i]
    return np.array([err1,err2])
```

```python
def gradientDescent(X,Y,learningRate,maxItr):
    theta = np.array([0.0,0.0])
    grad = np.array([0.0,0.0])
    
    for i in range(maxItr):
        grad = gradient(theta,X,Y)
        theta[0] = theta[0] - learningRate*grad[0]
        theta[1] = theta[1] - learningRate*grad[1]
    return theta
```

```python
theta = gradientDescent(x,y,learningRate  = 0.001,maxItr = 500)
print(theta[0],theta[1])
```
