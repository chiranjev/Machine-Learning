```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
```

```python
mnist = load_digits()
X = mnist.data
Y = mnist.target
print(X.shape)
print(Y.shape)
```

```python
from sklearn.datasets import load_boston,make_classification,make_regression,make_blobs
```

```python
boston = load_boston()
```

```python
XB = boston.data
YB = boston.target
print(XB.shape)
print(YB.shape)
```

```python
X,Y = make_classification(n_samples = 100,n_features = 2,n_informative = 2,n_redundant = 0,n_classes = 2)
plt.scatter(X[:,0],X[:,1],c=Y,edgecolor='k')
plt.show()
```

```python
X,Y = make_classification(n_samples = 150,n_features = 2,n_informative = 2,n_redundant = 0,n_classes = 2,n_clusters_per_class = 2)
plt.scatter(X[:,0],X[:,1],c=Y,edgecolor='k')
plt.show()
```

```python
X,Y = make_classification(n_samples = 100,n_features = 2,n_informative = 2,n_redundant = 0,n_classes = 2)
plt.figure(0,figsize = (8,8))

plt.subplot(221)
plt.scatter(X[:,0],X[:,1],c=Y,edgecolor='k')

plt.subplot(222)
X1,Y1 = make_classification(n_samples = 100,n_features = 2,n_informative = 2,n_redundant = 0,n_classes = 2,n_clusters_per_class = 2)
plt.scatter(X1[:,0],X1[:,1],edgecolor = 'k')

plt.subplot(223)
plt.title("Blobs")
X2,Y2 = make_blobs(n_samples = 100,n_features = 2,centers = 3)
plt.scatter(X2[:,0],X2[:,1],edgecolor = 'k')
plt.subplot(224)
X3,Y3 = make_regression(n_features = 1,n_samples = 200,bias = 4, noise = 10)
plt.scatter(X3,Y3,color = 'orange',edgeColor = 'k')

plt.show()
```
