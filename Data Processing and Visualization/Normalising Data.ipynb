```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfx = pd.read_csv('datasets/linearX.csv')
dfy = pd.read_csv('datasets/linearY.csv')

x = dfx.values
y = dfy.values

# normalised data -> mean = 0 and standard deviation = 1
x2 = (x-x.mean())/x.std()

plt.scatter(x,y,label = "Raw Data")
plt.scatter(x2,y,label = "Normalised Data")
plt.legend()
plt.plot()
```
