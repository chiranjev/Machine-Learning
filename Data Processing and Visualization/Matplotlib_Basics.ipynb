## Plots using Matplotlib for Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,10,12,15])
plt.plot(x)
y = np.array([10,8,6,4,2,1])
plt.plot(y)
```

```python
x = np.array([1,2,3,4,5,10,12,15])
plt.plot(x)
plt.show()
y = np.array([10,8,6,4,2,1])
plt.plot(y)
plt.show()
```

```python
x = np.array([1,3,5,7,9])
y = np.array([100,80,70,90,100])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,x**2,label = '2017')
plt.plot(x,x**3,label = '2018')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y)
```

```python
plt.title("Some random data")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.scatter(x,y,marker = '^',color = 'red')
plt.show()
```

## Multivariate Normal Distribution

```python
# Bivariate
mean = np.array([0.0,0.0])
cov = np.array([[1,0],[0,1]]) # y is independent of change in x

dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],[0.8,1]]) # positive covariance -> as x increases y increases 

dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,-0.8],[-0.8,1]]) # negative covariance -> as x increases y decreases 

dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],[0.8,1]])
mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],[0.2,1.1]])

dist = np.random.multivariate_normal(mean,cov,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
